```python
def get_all_video_in_channel(channel_id):
    api_key = "AIzaSyCl2TSLxpC7ZLswmLrobD6a0atZMjJ63ns" #My API Key

    base_search_url = 'https://www.googleapis.com/youtube/v3/search?'

    first_url = base_search_url+'key={}&channelId={}&part=snippet,id&maxResults=25'.format(api_key, channel_id)

    videos_ids = []
    url = first_url
    while True:
        inp = urllib.request.urlopen(url)
        resp = json.load(inp)

        for item in resp['items']:
            if item['id']['kind'] == "youtube#video":
                videos_ids.append(item['id']['videoId'])
        try:
            next_page_token = resp['nextPageToken']
            url = first_url + '&pageToken={}'.format(next_page_token)
        except:
            break
    return videos_ids
```

```python
def get_all_playlists_in_channel(channel_id):
    api_key = "AIzaSyCl2TSLxpC7ZLswmLrobD6a0atZMjJ63ns" #My API Key

    base_search_url = 'https://www.googleapis.com/youtube/v3/search?'

    first_url = base_search_url+'key={}&channelId={}&part=snippet,id&maxResults=25'.format(api_key, channel_id)

    videos_ids = []
    url = first_url
    while True:
        inp = urllib.request.urlopen(url)
        resp = json.load(inp)

        for item in resp['items']:
            if item['id']['kind'] == "youtube#playlist":
                videos_ids.append(item['id']['playlistId'])
        try:
            next_page_token = resp['nextPageToken']
            url = first_url + '&pageToken={}'.format(next_page_token)
        except:
            break
    return videos_ids
```

```python
def get_all_video_in_channel_in_playlist(list_ID):
    api_key = "AIzaSyCl2TSLxpC7ZLswmLrobD6a0atZMjJ63ns" #My API Key


    first_url = "https://www.googleapis.com/youtube/v3/playlistItems?part=snippet,id&maxResults=25&playlistId={}&key={}".format(list_ID, api_key)

    videos_ids = []
    url = first_url
    while True:
        inp = urllib.request.urlopen(url)
        resp = json.load(inp)

        for item in resp['items']:
            if item['snippet']['resourceId']['kind'] == "youtube#video":
                videos_ids.append(item['snippet']['resourceId']['videoId'])
        try:
            next_page_token = resp['nextPageToken']
            url = first_url + '&pageToken={}'.format(next_page_token)
        except:
            break
    return videos_ids
```

```python
def calcDuration(time):
    seconds = 0;
    time = time[2:]
    num = ""
    for i in time:
        if ord(i) >= 48 and ord(i) <= 57:
            num+= i
        else:
            if(i == 'H'):
                seconds+= int(num)*60*60
            elif(i == 'M'):
                seconds+= int(num)*60
            else:
                seconds+= int(num)
            num = ""
    return seconds
```

```python
def get_all_details_about_video(video_ids):
    my_data_structure = {
        'video_Link': [],
        'publishedAt': [],
        'title': [],
        'description': [],
        'tags': [], # func
        'categoryId': [],
        'defaultLanguage': [],
        'defaultAudioLanguage': [],
        'definition' : [],
        'duration': [], #func
        'projection': [], #?
        'viewCount': [],
        'likeCount': [],
        'dislikeCount': [],
        'favoriteCount': [],
        'commentCount': [],
    }
    api_key = "AIzaSyCl2TSLxpC7ZLswmLrobD6a0atZMjJ63ns" #My API Key
    base_video_url = 'https://www.youtube.com/watch?v='
    for video_id in video_ids:
        video_url = "https://www.googleapis.com/youtube/v3/videos?id={}&key={}&part=snippet,contentDetails,statistics,status".format(video_id, api_key)
        inp = urllib.request.urlopen(video_url)
        resp = json.load(inp)
        try:
            itemsJSON = resp['items'][0];
            try:
                my_data_structure['video_Link'].append(base_video_url + video_id)
            except:
                my_data_structure['video_Link'].append('')
            try:
                my_data_structure['publishedAt'].append(itemsJSON['snippet']['publishedAt'])
            except:
                my_data_structure['publishedAt'].append('')
            try:
                my_data_structure['title'].append(itemsJSON['snippet']['title'])
            except:
                my_data_structure['title'].append('')
            try:
                my_data_structure['description'].append(itemsJSON['snippet']['description'])
            except:
                my_data_structure['description'].append('')
            try:
                tagsAsString = ",".join(itemsJSON['snippet']['tags'])
                my_data_structure['tags'].append(tagsAsString)
            except:
                 my_data_structure['tags'].append([])
            try:
                my_data_structure['categoryId'].append(itemsJSON['snippet']['categoryId'])
            except:
                 my_data_structure['categoryId'].append('')
            try:
                my_data_structure['defaultLanguage'].append(itemsJSON['snippet']['defaultLanguage'])
            except:
                my_data_structure['defaultLanguage'].append('')
            try:
                my_data_structure['defaultAudioLanguage'].append(itemsJSON['snippet']['defaultAudioLanguage'])
            except:
                my_data_structure['defaultAudioLanguage'].append('')
            try:
                my_data_structure['definition'].append(itemsJSON['contentDetails']['definition'])
            except:
                my_data_structure['definition'].append('')
            try:
                duration = calcDuration(itemsJSON['contentDetails']['duration'])
                my_data_structure['duration'].append(duration) #must 
            except:
                duration = calcDuration('')
            try:
                my_data_structure['projection'].append(itemsJSON['contentDetails']['projection'])
            except:
                my_data_structure['projection'].append('')
            try:
                my_data_structure['viewCount'].append(itemsJSON['statistics']['viewCount'])
            except:
                my_data_structure['viewCount'].append('')
            try:
                my_data_structure['likeCount'].append(itemsJSON['statistics']['likeCount'])
            except:
                my_data_structure['likeCount'].append('')
            try:
                my_data_structure['dislikeCount'].append(itemsJSON['statistics']['dislikeCount'])
            except:
                my_data_structure['dislikeCount'].append('')
            try:
                my_data_structure['favoriteCount'].append(itemsJSON['statistics']['favoriteCount'])
            except:
                my_data_structure['favoriteCount'].append('')
            try:
                my_data_structure['commentCount'].append(itemsJSON['statistics']['commentCount'])
            except:
                my_data_structure['commentCount'].append('')
        except:
            pass
    return my_data_structure
```
